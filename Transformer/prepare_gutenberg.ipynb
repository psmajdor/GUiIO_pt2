```python
import pandas as pd
import shutil
import os
from glob import glob
```

## Read 10k english books

```python
df = pd.read_csv("dataset/gutenberg_over_70000_metadata.csv")
english_books = df[df['Language'] == 'English']
english_books = english_books.head(20000)

os.makedirs("dataset/gutenberg_10k", exist_ok=True)

display(english_books.head(10))
```

```python
def count_files(root):
    files_count = sum(len(files) for _, _, files in os.walk(root))

    return files_count
```

```python
print_once = False

for index, row in english_books.iterrows():
    book_id = row['Book Num']
    book_path = glob(f"dataset/books_20000/{book_id}_*.pkl")
    dst_path = f"dataset/gutenberg_10k/{book_id}.pkl"

    if len(book_path) == 0:
        continue

    f_count = count_files("dataset/gutenberg_10k")

    if not print_once:
        print_once = True
        print(f_count)

    if f_count >= 10000:
        break

    if os.path.exists(book_path[0]):
        shutil.move(book_path[0], dst_path)
```
